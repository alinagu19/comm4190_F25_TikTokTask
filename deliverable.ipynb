## TikTok Task: Comment Analysis Writeup


### Video URL and Description

Video URL: https://www.tiktok.com/t/ZP8DppJdS/ 

Description: The video is of a performance from this season of Dancing with the Stars featuring Whitney Leavitt (celebrity) and Mark Ballas. They are performing a tango to “Golden” from KPop Demon Hunters and there are a lot of mixed opinions of the dance and also Whitney since many people believe that she doesn’t deserve to be on the show because of her extensive dance experience. 

### Research Question

Do people like Whitney and her dances?


### Concept and Coding Changes

concept1 =  Code 1 talks positively or neutral about Whitney
                Leavitt, her partner, or her dance, Code 0 if the 
                comment is unrelated or negative
            

concept2 =  Code 1 talks negatively or neutral about Whitney 
                Leavitt, her partner, or her dance, Code 0 if the
                comment is unrelated or positive

We ended up having to change the concept prompts to include the neutral and unrelated cases in order to be able to classify all possible comments on the video. We also added the inclusion of topics including Whitney’s partner and her dance which were related to her but did not necessarily directly have opinions about her.

### Final LLM Prompt and Interrater Reliability Scores

LLM_prompt = You are a research assistant. You will receive a series of tiktok comments, please code for the sentiment of how the user felt about Whitney Leavitt. You will receive {len(text_list)} texts.
        
                Return ONLY valid JSON — a list of {len(text_list)}                 
                dictionaries like this:
        
                [
                  {{"concept1": 1, "concept2": 0}},
                  ...
                ]
        
                No explanations, no comments, no extra text — only the 
                JSON array.
                Texts:
                {texts_formatted}

             
Krippendorff’s Alpha between Human and AI:

• Concept 1: 0.704

• Concept 2: 0.528


These values indicate moderate agreement meaning better than chance and reasonably consistent, but still leaving room for disagreement and ambiguity in how some comments were interpreted

### Prompting Strategies Reflection

We kept our LLM prompt pretty simple but mostly refined our concepts to be more descriptive to give the LLM context for what our coding task was. This better defined where positive, negative, and neutral comments would fall into the coding schema. By tightening the prompt to explicitly say that the goal was to code “how the user felt about Whitney Leavitt,” we focused the LLM on her and not about Dancing with the Stars or the dancing in general. We believe these changes reduced random variation in the LLM’s responses and made its coding behavior more predictable, which is reflected in the moderate Krippendorff’s alpha scores we ultimately achieved for both concepts.

### Results Write-up
Overall, in this task we were trying to determine whether or not the general public liked Whitney and her dances because TikTok seems to be quite split on opinions toward her. The research question was “Do people like Whitney and her dances?” In total, we had two concepts with the first being “Code 1 talks positively or neutral about Whitney Leavitt, her partner, or her dance, Code 0 if the comment is unrelated or negative” and the second being “Code 1 talks negatively or neutral about Whitney Leavitt, her partner, or her dance, Code 0 if the comment is unrelated or positive.” We then coded 100 comments as a team and then compared our agreeability with the LLM. For concept 1, the reliability score was 0.704 and for concept 2, the reliability score was 0.528. This implies that concept 1 may have been more clearly defined or that the comments indicating positivity were much more concretely positive. Comments that we rated as 1 for concept 2 were generally more ambiguous or required context. Additionally, many would likely be deemed as “TikTok Lingo”, which would make it more difficult for an LLM to discern. 

Overall, concept 1 (which praised Whitney and the dance) received roughly 67% more 1 scores, which indicates that the comments leaned more positive. However, something really interesting to note was that although there were less comments labeled as concept 2, they tended to receive more likes with the ratio of concept 1 to concept 2 likes being about 1:3. The average likes per comment type also supports this with the average concept 2 comment having about 3x more likes than the average concept 1 comment. 

Something that might explain this result is the fact that this video was taken directly from the Dancing with the Stars Tiktok account rather than a fanpage. As a result, this video was fairly unbiased and the viewers were more likely to be casual watchers of the show rather than people with strong negative opinions. However, those with negative opinions may be less likely to comment on the official page since their profile is directly attached to the comment. As such, they’re more likely to just like/upvote comments that share the same sentiment. The positive comments are much more superficial and wouldn’t receive any backlash. These results may not outright give a robust answer towards our research question, but it does reveal that negative comments toward Whitney are more popular, which reveals the sentiment of many viewers. She definitely does not have the same popularity of other contestants on the season, and this proved to be detrimental as she was eliminated during the semi-finals.
